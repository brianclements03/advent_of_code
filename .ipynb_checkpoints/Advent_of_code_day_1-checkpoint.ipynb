# Part 2
Your goal now is to count the number of times the sum of measurements in this sliding window increases from the previous sum. So, compare A with B, then compare B with C, then C with D, and so on. Stop when there aren't enough measurements left to create a new three-measurement sum.

```python
# 199  A      
# 200  A B    
# 208  A B C  
# 210    B C D
# 200  E   C D
# 207  E F   D
# 240  E F G  
# 269    F G H
# 260      G H
# 263        H
```

```python
output2 = []
for i, row in depth.iterrows():
    if i > 0 :
        if i < (len(depth)-2):
            if ((depth.iloc[i]['depth_diff'] + depth.iloc[i+1]['depth_diff'] 
                + depth.iloc[i+2]['depth_diff']) > (depth.iloc[i-1]['depth_diff'] 
                + depth.iloc[i]['depth_diff'] + depth.iloc[i+1]['depth_diff'])):
                    output2.append(1)
            else:
                continue
```
